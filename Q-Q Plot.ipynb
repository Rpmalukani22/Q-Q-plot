# Q-Q plot example

```python
import pylab
import scipy.stats as stats
import numpy as np
```

```python
normal_dist = np.random.normal(loc=0,scale=1,size=1000000) #try different sizes ! -Ruchitesh
uniform = np.random.uniform(low=1,high=6,size=50000)#try different sizes ! -Ruchitesh
stats.probplot(normal_dist,dist="norm",plot=pylab)
pylab.show()
stats.probplot(uniform,dist="norm",plot=pylab)
pylab.show()
```

#### Observation
Here in the first plot we can see we can see best fitted line. Thus both the distributions are same. The opposite behaviour can be seen in the second graph.

#### Disadvantage

- If sample size is small then it is difficult to interpret the plot
