# Q-Q plot example

```python
import pylab
import scipy.stats as stats
import numpy as np
```

```python
normal_dist = np.random.normal(loc=0,scale=1,size=1000000) #try different sizes !
uniform = np.random.uniform(low=1,high=6,size=50000)#try different sizes !
stats.probplot(normal_dist,dist="norm",plot=pylab)
pylab.show()
stats.probplot(uniform,dist="norm",plot=pylab)
pylab.show()
```
